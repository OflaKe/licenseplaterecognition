# 6. Load Custom Model
#aici rulezi modelul in sine comanda de mai jos da run direct, urmatoarea e pentru run la "algoritm" si abia a 3-a ar fii fost 

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp5/weights/best.pt', force_reload=True) 
```

```python
IMAGES_PATH = os.path.join('dataset' , 'images' )
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to read frame")
        break

    # Resize the frame
    #frame_resized = cv2.resize(frame)
    backup_frame =frame.copy()
    results =  model(frame)
    # Show the frame on screen
    cv2.imshow('XD', np.squeeze(results.render())) # Replace np.squeeze(results.render()) if you have detection results.

    # Handle key press events
    key = cv2.waitKey(10) & 0xFF
    if key == ord('q'):  # Quit the program
        print("Exiting...")
        break
    elif key == ord('s'):  # Save the image
        imgname = os.path.join(IMAGES_PATH, f"{uuid.uuid1()}.jpg")
        
        #cv2.imshow('XD', backup_frame)
        cv2.imwrite(imgname, backup_frame)
        print(f"Image saved: {imgname}")

#Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    frame_resized = cv2.resize(frame, (1024, 720))
    # Make detections 
    results = model(frame_resized)
    #print(results)
    cv2.imshow('XD', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
